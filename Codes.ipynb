```python
import pandas as pd
import numpy as np
covid_19_india=pd.read_csv('covid_19_india.csv')

covid_19_india

```

```python
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
```

```python
covid_19_india.rename(columns={'State/UnionTerritory': 'State', 'Cured': 'Recovered'}, inplace=True)
covid_19_india
```

```python
total_data = covid_19_india[covid_19_india['Date'] ==  "2021-08-11"]
total_data
```

```python


# Calculate CFR for each state
total_data['CFR'] = (total_data['Deaths'] / total_data['Confirmed']) * 100

# Display the state-wise CFR
cfr_data  = total_data[['State', 'CFR']].sort_values(by='CFR', ascending=False)
cfr_data
#convert dataframe into csv

cfr_data.to_csv('cfr_data.csv', index=False)

```

```python
# Read data from the CSV files
hospital_beds_df = pd.read_csv("Hospital_Beds - Sheet1.csv")
covid_df = total_data

# Select the required columns from each DataFrame
hospital_beds_subset = hospital_beds_df[['State', 'Total_Population_in_thousands']]
covid_subset = covid_df[['State', 'Confirmed']]

# Merge the data based on the common "State" column
merged_df = pd.merge(hospital_beds_subset, covid_subset, on='State')

# Display the resulting DataFrame
print(merged_df)
```

```python
merged_df['cases_per_population'] = (merged_df['Confirmed']*100)/(merged_df['Total_Population_in_thousands']*1000)
merged_df
```

```python
cfr_data  = merged_df[['State', 'cases_per_population']].sort_values(by='cases_per_population', ascending=False)
cfr_data
#convert dataframe into csv

cfr_data.to_csv('cases_per_population.csv', index=False)
```

## kerala graphs 

```python
kerala_data = covid_19_india[covid_19_india['State']== 'Kerala']
kerala_data
```

```python
kerala_data.drop(columns={'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'}, inplace=True)
```

```python
kerala_data
```

```python
import pandas as pd


kerala_data['Date'] = pd.to_datetime(kerala_data['Date'])

# Sort DataFrame by date
kerala_data = kerala_data.sort_values(by='Date')

# Calculate the difference in cases between consecutive days
kerala_data['New_Recovered'] = kerala_data['Recovered'].diff().fillna(kerala_data['Recovered'])
kerala_data['New_Deaths'] = kerala_data['Deaths'].diff().fillna(kerala_data['Deaths'])
kerala_data['New_Confirmed'] = kerala_data['Confirmed'].diff().fillna(kerala_data['Confirmed'])

# Display the DataFrame with new columns
kerala_data
```

```python
import matplotlib.pyplot as plt

# Plot for New_Deaths vs Date
plt.figure(figsize=(10, 6))
plt.plot(kerala_data['Date'], kerala_data['New_Deaths'], color='red')
plt.title('No. of death vs Date')
plt.xlabel('Date')
plt.ylabel('No. of Deaths')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plot for New_Recovered vs Date
plt.figure(figsize=(10, 6))
plt.plot(kerala_data['Date'], kerala_data['New_Recovered'], color='green')
plt.title('Recovery vs Date')
plt.xlabel('Date')
plt.ylabel('Number of people recovered')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plot for New_Confirmed vs Date
plt.figure(figsize=(10, 6))
plt.plot(kerala_data['Date'], kerala_data['New_Confirmed'], color='blue')
plt.title('Total new cases vs Date')
plt.xlabel('Date')
plt.ylabel('Total new positive cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the CFR CSV file into a pandas DataFrame
cfr_df = pd.read_csv("cfr_data.csv")

# Load and merge each of the other CSV files with the CFR DataFrame
files = ["Health_Index - Sheet1.csv", "beds.csv", "Average_age - Sheet1.csv", "per_capita_income - Sheet1.csv", "population_index - Sheet1 (1).csv","cases_per_population.csv"]
correlation_coefficients = {}

for file in files:
    # Load the other CSV file into a pandas DataFrame
    other_df = pd.read_csv(file)
    
    # Merge the other DataFrame with the CFR DataFrame on the "State" column
    merged_df = pd.merge(cfr_df, other_df, on="State")
    
    # Calculate the Pearson correlation coefficient
    correlation_coefficient = merged_df["CFR"].corr(merged_df.iloc[:, 2])
    
    # Save the correlation coefficient for the current variable
    variable_name = file.split(".")[0]  # Extract variable name from file name
    correlation_coefficients[variable_name] = correlation_coefficient

# Print the correlation coefficients
for variable, coefficient in correlation_coefficients.items():
    print(f"Pearson correlation coefficient between CFR and {variable}: {coefficient}")
```

```python
import pandas as pd

# Load the CFR and beds data
cfr_df = pd.read_csv("cfr_data.csv")
beds_df = pd.read_csv("beds.csv")

# Merge the CFR and beds dataframes based on the common "State" column
merged_df = pd.merge(cfr_df, beds_df, on="State")

# Sort the merged dataframe by CFR values in ascending order
merged_df = merged_df.sort_values(by="CFR")

# Select the last 10 states with the least CFR values
last_10_states = merged_df.head(10)

# Calculate the average number of beds required for the last 20 states
avg_beds_required = last_10_states["Beds_per_thousand"].mean()

print("Average number of beds required for the last 20 states with the least CFR values:")
print(avg_beds_required)
```

```python
x = pd.read_csv('Hospital_Beds - Sheet1.csv')
x = x[['State','Total_Population_in_thousands']]
merged_df = pd.merge(merged_df,x, on="State")
# Sort the merged dataframe by CFR values in descending order
merged_df = merged_df.sort_values(by="CFR", ascending=False)

# Select the top 20 states with the highest CFR values
top_20_states = merged_df.head(20)


# Calculate the difference between the CFR of each state in the top 20 states and the average number of beds required
differences = {}
for index, row in top_20_states.iterrows():
    state = row["State"]
    cfr_value = row["Beds_per_thousand"]
    pop = row["Total_Population_in_thousands"]
    difference = (avg_beds_required - cfr_value )* pop
    
    differences[state] = difference

# Print the differences for each state
print("Difference between the CFR of each state in the top 20 states and the average number of beds required:")
for state, difference in differences.items():
    print(f"{state}: {difference}")
```
